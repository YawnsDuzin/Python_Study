###제목(Header)

---

### 2. 강조(Emphasis)
- 이탤릭체(Italic)
  - 텍스트 앞뒤로, ```*``` 또는 ```_```기호를 사용합니다.
- 볼드체(Bold)
  - 텍스트 앞뒤로 ```**```또는 ```__```기호를 사용합니다.
- 이탤릭체(Italic) 와 볼드체(Bold) 동시 적용
  - 텍스트 앞뒤로 ```**_```또는 ```__*```기호를 사용합니다.
---

*이탤릭체* 또는 _이탤릭체_

**볼드체** 또는 __볼드체__

**_볼드체와 이탤릭체_** 또는 __*볼드체와 이탤릭체*__

###3. 목록(Lists)
- 순서가 없는 목록을 나열할 때는 항목 앞에 ```-, *``` 또는 ```+```기호를 사용하여 목록을 생성합니다.
- 순서가 있는 항목을 나열할 때는 항목 앞에 ```숫자와 점(.)```을 사용하여 목록을 생성합니다.
---

순서 없는 목록 (Unordered Lists)
- 항목1
* 항목2
+ 항목3

순서 있는 목록(Ordered Lists)
1. 첫 번째 항목
2. 두 번째 항못
3. 세 번째 항목

상위/하위항목 목록
- 항목1
  - 항목1.1
  - 항목1.2
  - 항목1.3
- 항목2
  - 항목2.1
  - 항목2.2

###4. 수평선(Horizontal Line)
- 수평선을 삽입하려면 ```3개 이상의 하이픈(-), 별표(*), 또는 언더스코어(_)```를 연속으로 작성합니다. 이 세 가지 기호 중 어느 것을 사용해도 동일한 수평선이 생성됩니다.
---

---
구분하기(---)
***
구분하기(***)
___
구분하기(___)
