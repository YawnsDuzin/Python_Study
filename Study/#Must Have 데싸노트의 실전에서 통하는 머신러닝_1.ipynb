## 데이터 시각화그래프

```python
import seaborn as sns
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.lineplot(x = 'size', y = 'total_bill', data = data, hue = 'sex')
```

```python
sns.scatterplot(x = 'tip', y = 'total_bill', data = data, hue = 'sex')
```

```python
sns.histplot(data['size'])
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = data, hue = 'day')
```

```python
data.head()
data.info()
```

```python
sns.heatmap(data[['total_bill', 'tip', 'size']].corr())
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = data, hue = 'day')
```
